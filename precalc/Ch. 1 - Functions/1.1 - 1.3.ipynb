## 1) Complete the code to draw a full circle.

```python
# The graph is drawn from left to right, so a smaller input would show the graph as a
# zig-zag in the shape of a circle
temp = np.linspace(-1, 1, 1000)

def circle(x):
    return np.sqrt(1 - x**2)

x = []
y = []

for i in temp:
    # since y = sqrt(x) can be either y or -y, evaluate for both
    x.append(i)
    x.append(i)
    y.append(circle(i))
    y.append(-circle(i))

# The default graph dimensions show the graph as an ellipse, so resize it
plt.figure(figsize=(6, 6))
plt.plot(x, y)
plt.show()
```

#### Use this function to compute the rate of change of f = x^2 in the domain [-2, 2]
#### Plot the function and the rate of change on the same graph. 

#### Use this function to compute the rate of change of f = x in the domain [-2, 2]
#### Plot the function and the rate of change on the same graph. 

```python
def rate_of_change(f, step, domain):
    return ((f(domain + step) - f(domain))/step)
    
def f1(x):
    return x**2

def f2(x):
    return x

def f3(x):
    return x**3

domain = np.linspace(-2, 2, 100)

plt.plot(domain, f1(domain))
# since 5/100 is 0.05, 0.15 gives the average rate of change over 4 values in the domain
plt.plot(domain, rate_of_change(f1, 0.15, domain))
plt.show()

plt.plot(domain, f2(domain))
plt.plot(domain, rate_of_change(f2, 0.15, domain))
plt.show()

plt.plot(domain, f3(domain))
plt.plot(domain, rate_of_change(f3, 0.15, domain))
plt.show()
```
