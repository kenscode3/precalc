### Linear Equation from the Graph of a Line
When we have the graph of a line, we can find the equation for it. We first choose two points, then find the **rise** (how far up we must go to reach it), and the **run** (how far to the right we must go to reach it). For any two points we get a fraction that simplifies to $\frac{2}{1} = 2$ (which means $m = \frac{rise}{run}$). This tells use that the slope is 2, and we the find the y-intercept where the line crosses the y axis ($4$). We can then substitute these into the slope-intercept form to get the original equation for the line: $y = 2x + 4$.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()

def line(x):
    return 2 * x + 4

x = np.linspace(-10, 10, 21)

plt.ylim(-6, 6)
plt.xlim(-7, 7)
plt.plot(x, line(x))
```

### X-Intercept of a Line
We can find the x-intercept of a line by simply solving the linear equation for $x$ when $y = 0$:

$0 = 3x - 6$

$6 = 3x$

$\frac{6}{3} = \frac{3x}{3}$

$2 = x$

So the x-intercept for the linear equation $y = 3x - 6$ is $2$.

### Horizontal and Vertical Lines
A horizontal line is a line defined by an equation in the form $f(x) = b$ (it's a constant function).

A vertical line is a line defined by the equation in the form $x = a$.

### Determining Whether Lines are Parallel or Perpendicular
We can determine if two lines are parallel if their slopes are the same. For instance, $y = 2x + 3$ and $y = 2x - 6$ are parallel because they have the same slope ($m = 2$).

We can find whether two lines are perpendicular or not by using one of two methods:

1) The product of the slopes

If the product of the slopes is $-1$, then the lines are perpendicular.

2) Negative reciprocal

If the slope of one line is equal to the negative reciprocal of the other, then the lines are perpendicular.

### Finding Equations for Parallel and Perpendicular Lines to a Given Line
We know that lines are parallel by simply comparing the slopes. We can also find the equation for a parallel line that passes through a particular point. For example, if we have the equation $y = 3x + 6$ and we want to find a parallel line that passes through (1, 7), then we can substitue this information into the point-slope form:

$y - y_1 = m(x - x_1)$

$y - 7 = 3(x - 1)$, which simplifies to:

$y = 3x + 4$

We can see on the graph below that the line passes though the point (1, 7).

```python
def line(x):
    return 3 * x + 6

def parallel(x):
    return 3 * x + 4

x = np.linspace(-10, 10, 21)

plt.ylim(-8, 8)
plt.xlim(-7, 7)
plt.plot(x, line(x))
plt.plot(x, parallel(x))
```

To find a line that is perpendicular to another line, we simply find the negative reciprocal of the given line's slope. If we want to find a perpendicular line that passes through a specific point we can subsitute the $x$ and $y$ of the point into the equation for the first line, which we'll define as $y = 2x + 4$. We then substitute the negative reciprocal of the original slope, then solve for $b$. We'll use the point (4, 0) as an example:

$0 = -\frac{1}{2}4 + b$

$0 = -2 + b$

$b = 2$

```python
def line(x):
    return (2 * x) + 4

def perpendicular(x):
    return ((-(1/2)) * x) + 2

x = np.linspace(-8, 8, 21)

# It's important to note that since the line(x) function returns higher values,
# it distorts the graph and it doesn't look perpendicular
plt.plot(x, line(x))
plt.plot(x, perpendicular(x))
```

### Solving Systems of Linear Equations Using a Graph
A system of linear equations includes two or more linear equations. The graphs of two lines will intersect if they are not parallel. If two parallel lines are the same they are **coincident**, which means they intersect at every point. By solving two functions for $x$, we can find the input where two linear functions intersect.

If we have $f(x) = 2x + 3$ and $g(x) = 3x - 2$, then we first solve for $x$:

$2x + 3 = 3x - 2$

$3 + 2 = 3x - 2x$

$5 = x$

So now we know that the two linear functions intersect when the input is $5$. We can then evaluate either function with the input of $5$ to get the y value of the point:

$y = 3(5) - 2$

$y = 13$

So the point where these two linear functions intersect is $(5, 13)$.

We can see on the graph below that this is correct.

```python
def f(x):
    return 2 * x + 3

def g(x):
    return 3 * x - 2

x = np.linspace(-6, 6, 21)

plt.xlim(4, 6)
plt.ylim(10, 15)
plt.plot(x, f(x))
plt.plot(x, g(x))
```
