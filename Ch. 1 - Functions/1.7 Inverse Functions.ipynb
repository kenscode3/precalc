### Inverse Functions
For any one-to-one function $f(x) = y$, a function $f^{-1}(x)$ is an **inverse function** of $f$ if $f^{-1}(y) = x$. This can also be written as $f^{-1}(f(x)) = x$ for all $x$ in the domain of $f$. It also follows that $f(f^{-1}(x)) = x$ for all $x$ in the domain of $f^{-1}$ if $f^{-1}$ is the inverse of $f$.

The notation $f^{-1}$ is read "$f$ inverse." Like any other function, any variable name can be used as the input for $f^{-1}$, so it can be written $f^{-1}(x)$, which reads as "$f$ inverse of $x$." Note that $f^{-1}(x) \neq \frac{1}{f(x)}$

Not all functions have inverses, because some outputs share an input.

##### Example
If $f(x) = 2x$, then $f^{-1}(x) = \frac{x}{2}$

$f(6) = 2(6) = 12$, then $f^{-1}(12) = \frac{12}{2} = 6$

### Testing Inverse Functions Algebraically
Given two functions $f(x)$ and $g(x)$, testing if they are inverses of each other is done by evaluating $f(g(x))$ and $g(f(x))$ for $x$. If both functions equal $x$, then they are inverses of each other. If one or niether functions equal $x$, then they are not inverse.

### Domain and Range of Inverse Functions
The range of a function $f(x)$ is the domain of the inverse function $f^{-1}(x)$. The domain of $f(x)$ is the range of $f^{-1}(x)$.

Given a function, to find the domain and range of its:

- If the function is one-to-one, write the range of the original function as the domain of the inverse, and vice-versa for the inverse.

- If the function is not one-to-one, then the restricted domain of the original function becomes the range of the inverse function.

### Evaluating a Function and Its Inverse from a Graph
Given a graph, we can find $g(x)$ and $g^{-1}(x)$. For the graph of a $f(x) = 2x$, if $x = 5$, we find that the output is $10$. The inverse of $g(x)$ is the point on the graph where the ouput is equal to the input of the original function ($5$), which is $2.5$.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()

x = np.linspace(-10, 10, 5)

def f(x):
    return 2 * x

plt.plot(x, f(x))
```

### Finding Inverses of Functions Represented by Formulas
Sometimes it's necessary to know an inverse function for all elements, not just a few. By solving the equation of the function for x we can show the inverse function for the entire range of $f(x)$. For the simple function $f(x) = 2x$ we have:

$2x = y$

$\frac{2x}{2} = \frac{y}{2}$

$x = \frac{y}{2}$

So, $f^{-1}(x) = \frac{x}{2}$

### Graphing Inverse functions

```python
x = np.linspace(-2, 2, 51)

def g(x):
    return 2 * x

def g_inverse(x):
    return x / 2

plt.plot(x, g(x))
plt.plot(x, g_inverse(x))
```
