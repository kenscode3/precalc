### Absolute Value Functions
The distance of two numbers can be determined by $|x - n|$. This is useful when you want to describe the values that fall within a certain range of a number. For instance $|x - 5| \leq 4$ describes all values that are within $\pm4$ of $5$.

Absolute value functions can be solved algebraically:

$x - 5 = 4$ or $x - 5 = -4$

$x - 5 + 5 = 4 + 5$

$x = 9$

and,

$x - 5 = -4$

$x - 5 + 5= -4 + 5$

$x = 1$

This gives us the interval $[1, 9]$ or the inequality $1 \leq x \leq 9$.

#### Solutions to Absolute Value Equations

For real numbers $A$ and $B$, and equation of the form $|A| = B$, with $B \geq 0$, will have solutions when $A = B$ or $A = -B$. If $B < 0$, the equation $|A| = B$ has no solution.

### Solving an Absolute Value Inequality Using a Graph
A graph can be used to find the interval for $|x - 5| \leq 4$. By graphing the equation and finding the points where $y = 4$ we can find the interval that is true for the equation. Since the graph intersects the line at $1$ and $9$, the interval is $[1, 9]$

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()

def f(x):
    return abs(x - 5)

x = np.linspace(0, 10, 21)
y = [4] * 21

plt.plot(x, y)
plt.plot(x, f(x))
```
