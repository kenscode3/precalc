### Vertical Shifts
Given a function $f(x)$, a new function $g(x) = f(x) + k$, is a vertical shift of the function $f(x)$.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return x**2

def g(x):
    return f(x) + 3

x = np.linspace(-1, 1, 10)

plt.plot(x, f(x))
plt.plot(x, g(x))
```

### Horizontal Shifts
Given a function $f(x)$, a new function $g(x) = f(x + n)$, is a horizontal shift of the function $f(x)$.

```python
def f(x):
    return x**2

def g(x):
    return f(x + 5)

x = np.linspace(-10, 10, 100)

plt.plot(x, f(x))
plt.plot(x, g(x))
```

### Combining Horizontal and Vertical Shifts

```python
def f(x):
    return x**2

def g(x):
    return f(x + 5) + 50

x = np.linspace(-10, 10, 100)

plt.plot(x, f(x))
plt.plot(x, g(x))
```

### Reflections

Given a function $f(x)$, the function $g(x) = -f(x)$ is a **vertical reflection** (flips the graph over the x-axis).

```python
def f(x):
    return abs(x)

x = np.linspace(-10, 10, 101)

# the -f(x) function call can be considered g(x) = -f(x)
plt.plot(x, -f(x))
plt.plot(x, f(x))
```

Given a function $f(x)$, the function $g(x) = f(-x)$ is a **horizontal reflection** (flips the graph over the y-axis).

```python
def f(x):
    return x**3

x = np.linspace(-10, 10, 101)

# the f(-x) function call can be considered g(x) = f(-x)
plt.plot(x, f(-x))
plt.plot(x, f(x))

def g(x):
    return 2**x
```

### Even and Odd Functions
A function that is symmetric about the y-axis (e.g., $f(x) = x^2$, and $f(x) = |x|$) is called an **even function**.

A function that is symmetric about the origin (e.g., $f(x) = x^3$) is called an **odd function**.

A function can be neither even nor odd if it is not symmetric (e.g., $f(x) = 2^x$).

The only function that is even *and* odd is the constant function $f(x) = 0$

#### Determining if a Function is Even, Odd, or Neither
By determining if a function satisfies certain conditions, it can be shown to be even, odd, or neither:

If $f(x) = f(-x)$, it is even.

If $(fx) = -f(-x)$, it is odd.

If does not satisfy either rule, it is neither.

### Stretches and Compressions
The vertical stretch or vertical compression of function $f(x)$ is a new function $g(x) = af(x)$, where $a$ is a constant. If $a$ is a non-zero real number, the absolute value of $a$ gives the amount of compression, and the sign shows whether the function is vertically reflected or not.

```python
def f(x):
    return abs(x)

def g(x):
    return 2 * f(x)

def h(x):
    return -0.25 * f(x)

x = np.linspace(-10, 10, 51)

plt.plot(x, f(x))
plt.plot(x, g(x))
plt.plot(x, h(x))
```

Function $f(x)$ can be horizontally stretched and compressed by multiplying the function's input by a positive constant ($f(ax)$). Multiplying the input by a negative constant will horizontally reflect the function and possibly compress or stretch it. The stretch or compression of the function can be remembered as $\frac{1}{a}$

```python
def f(x):
    return x**3

x = np.linspace(-10, 10, 51)

plt.plot(x, f(x))
plt.plot(x, f(x * -2))
plt.plot(x, f(x * 0.25))
```

### Sequences of Transformations
The order of operations matter dealing with transformations:

```python
def f(x):
    return x**2

x = np.linspace(-1, 1, 25)

# vertical transformations
# evaluate f(x), then shift 3, then stretch by 2
plt.plot(x, 2 * (f(x) + 3))
# evaluate f(x), then stretch by 2, then shift by 3
plt.plot(x, (2 * f(x)) + 3)

# horizontal transformations
# compress by 0.5 (1/2), then shift by 0.25
plt.plot(x, f((2 * x) + 0.25))
# shift by 0.25, then compress by 0.5 (1/2)
plt.plot(x, f(2 * (x + 0.25)))
```

Combining transformations is summarized as follows:

When combining vertical transformations written in the form $af(x) + k$, first vertically stretch by a then vertically shift by $k$.

When combining horizontal transformations written in the form $f(bx + h)$, first horizontally shift by h and the horizontally stretch by $\frac{1}{b}$

When combining horizontal transformations written in the form $f(b(x + h))$, first horizontally stretch by $\frac{1}{b}$ and then horizontally shift by $h$.

Horizontal and vertical transformations are independent. **Horizontal transformations are evaluated as the original function and vertical transformations are evaluated by a different function, after evaluating for the original function**.
